# Matplotlib 基础
在使用**Numpy**之前，需要了解一些画图的基础。

**Matplotlib**是一个类似Matlab的工具包，主页地址为

http://matplotlib.org

导入 matplotlib 和 numpy：

```python
%pylab
```

## plot 二维图

    plot(y)
    plot(x,y)
    plot(x, y, format_string)
 默认下标为`x`轴

```python
%matplotlib inline
x = linspace(0,2 * pi, 50)
plot(sin(x))
```

```python
plot(x,sin(x))
```

多条数据线:

```python
plot(x, sin(x), x, sin(2*x))
```

使用字符串给定线条参数:

```python
plot(x, sin(x), 'r-^')
```

```python
plot(x, sin(x), 'b-o', x, sin(2 * x), 'y-^')
```

## scatter 散点图

```python
scatter(x, sin(x))
```

```python
x = rand(200)
y = rand(200)
size = rand(200) * 30
color = rand(200)
scatter(x,y,size,color)
colorbar()
```

## 多图

使用figure()命令产生新的图像

```python

t = linspace(0, 2*pi, 50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
figure()
```

或者使用 subplot 在一幅图中画多幅子图：

    subplot(row, column, index)

```python
subplot(1,2,1)
plot(x)
subplot(1,2,2)
plot(y)
```
