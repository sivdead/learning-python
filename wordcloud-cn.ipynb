```python
filename = 'tensorflow.txt'
with open(filename, encoding='utf-8') as f:
    mytext = f.read()
import jieba
mytext = ' '.join(jieba.cut(mytext))
import numpy as np
from PIL import Image
from wordcloud import WordCloud,ImageColorGenerator
mask = np.array(Image.open("alice.png"))
image_colors = ImageColorGenerator(mask)
wordcloud = WordCloud(font_path="simsun.ttf", mask=mask,background_color="white").generate(mytext)
%pylab inline
wordcloud.to_file("tensorflow.png")
import matplotlib.pyplot as plt
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis("off")
plt.figure()
wordcloud.recolor(color_func=image_colors).to_file('alice-recolor.png')
```
